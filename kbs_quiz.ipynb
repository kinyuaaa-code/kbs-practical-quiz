```python
#!/usr/bin/env python
# coding: utf-8

# # Knowledge-Based System: Campus Access Control
# 
# **A Rule-Based System for University Facility Access**
# 
# **Student Name:** [Elvis Kinyua]
# **Reg Number:** [666842]
# **Date:** [15/02/2026]

# ## 1. Implement the KBS Engine

class InferenceEngine:
    """
    A forward-chaining knowledge-based system for making decisions
    based on facts and rules.
    """
    
    def __init__(self, system_name="Campus Access KBS"):
        self.system_name = system_name
        self.knowledge_base = set()      # Stores all known facts
        self.rule_set = []                # Stores all rules
        self.inference_log = []            # Tracks what was inferred
    
    def add_initial_fact(self, fact_statement):
        """Add a fact to start the knowledge base"""
        self.knowledge_base.add(fact_statement)
        print(f"âœ… INITIAL FACT LOADED: '{fact_statement}'")
        return self
    
    def define_rule(self, condition_function, resulting_fact, rule_description=""):
        """
        Add a production rule to the system
        """
        rule_entry = {
            "check": condition_function,
            "output": resulting_fact,
            "description": rule_description if rule_description else resulting_fact
        }
        self.rule_set.append(rule_entry)
        print(f"âš™ï¸ RULE DEFINED: IF {rule_entry['description']} THEN '{resulting_fact}'")
        return self
    
    def run_forward_chaining(self):
        """
        Execute the inference process until no new facts can be derived
        """
        print("\n" + "â•" * 65)
        print(f"ğŸš€ STARTING INFERENCE: {self.system_name}")
        print("â•" * 65)
        
        round_number = 1
        new_information_added = True
        
        while new_information_added:
            new_information_added = False
            print(f"\nâ–¶ï¸ INFERENCE ROUND {round_number}")
            print(f"   Current knowledge ({len(self.knowledge_base)} facts): {sorted(self.knowledge_base)}")
            
            for rule in self.rule_set:
                if rule['output'] in self.knowledge_base:
                    continue
                
                if rule['check'](self.knowledge_base):
                    self.knowledge_base.add(rule['output'])
                    self.inference_log.append(f"Round {round_number}: {rule['output']}")
                    print(f"   âœ¨ INFERRED: '{rule['output']}'")
                    new_information_added = True
            
            round_number += 1
        
        print("\n" + "â•" * 65)
        print("âœ… INFERENCE COMPLETE - Knowledge base saturated")
        print("â•" * 65)
        return self.knowledge_base
    
    def check_fact_exists(self, fact_to_check):
        return fact_to_check in self.knowledge_base
    
    def display_results(self):
        print("\n" + "â–ˆ" * 65)
        print("ğŸ“Š FINAL INFERENCE RESULTS")
        print("â–ˆ" * 65)
        
        print("\nğŸ“Œ NEW FACTS DISCOVERED:")
        if self.inference_log:
            for log_entry in self.inference_log:
                print(f"   â€¢ {log_entry}")
        else:
            print("   No new facts were discovered")
        
        print("\nğŸ“Œ COMPLETE KNOWLEDGE BASE:")
        for fact in sorted(self.knowledge_base):
            print(f"   â€¢ {fact}")
        
        return self.knowledge_base


# ## 2. Add Base Facts

print("\n" + "â–ˆ" * 65)
print("ğŸ”° STEP 1: INITIALIZING BASE FACTS")
print("â–ˆ" * 65)

campus_kbs = InferenceEngine("University Research Facility Access System")

campus_kbs.add_initial_fact("security_badge_issued_2025")
campus_kbs.add_initial_fact("research_personnel_status")
campus_kbs.add_initial_fact("safety_training_completed")
campus_kbs.add_initial_fact("temperature_screening_passed")
campus_kbs.add_initial_fact("equipment_certification_active")
campus_kbs.add_initial_fact("visitor_log_signed")

print(f"\nğŸ“‹ TOTAL BASE FACTS LOADED: {len(campus_kbs.knowledge_base)}")


# ## 3. Add Rules

print("\n" + "â–ˆ" * 65)
print("âš™ï¸ STEP 2: DEFINING RULES")
print("â–ˆ" * 65)

def has_fact(fact_set, required_fact):
    return required_fact in fact_set

# --- SINGLE CONDITION RULES ---
def rule_badge_check(facts):
    return has_fact(facts, "security_badge_issued_2025")

campus_kbs.define_rule(rule_badge_check, "identity_confirmed", "security_badge_issued_2025 is present")

def rule_training_check(facts):
    return has_fact(facts, "safety_training_completed")

campus_kbs.define_rule(rule_training_check, "safety_cleared", "safety_training_completed is present")


# --- MULTI-CONDITION RULES ---
def rule_research_access(facts):
    return (has_fact(facts, "research_personnel_status") and 
            has_fact(facts, "equipment_certification_active"))

campus_kbs.define_rule(rule_research_access, "lab_access_authorized", "research_personnel_status AND equipment_certification_active")

def rule_visitor_processing(facts):
    return (has_fact(facts, "visitor_log_signed") and 
            has_fact(facts, "temperature_screening_passed"))

campus_kbs.define_rule(rule_visitor_processing, "temporary_access_granted", "visitor_log_signed AND temperature_screening_passed")


# --- CHAINED RULES ---
def rule_building_entry(facts):
    return (has_fact(facts, "identity_confirmed") and 
            has_fact(facts, "safety_cleared"))

campus_kbs.define_rule(rule_building_entry, "general_building_access", "identity_confirmed AND safety_cleared")

def rule_restricted_area(facts):
    return (has_fact(facts, "general_building_access") and 
            has_fact(facts, "lab_access_authorized"))

campus_kbs.define_rule(rule_restricted_area, "research_lab_entry_permitted", "general_building_access AND lab_access_authorized")

def rule_full_campus_access(facts):
    return (has_fact(facts, "research_lab_entry_permitted") or 
            has_fact(facts, "temporary_access_granted"))

campus_kbs.define_rule(rule_full_campus_access, "campus_entry_approved", "(research_lab_entry_permitted OR temporary_access_granted)")


# ## 4. Run Inference and Print Results

print("\n" + "â–ˆ" * 65)
print("ğŸ” STEP 3: RUNNING INFERENCE ENGINE")
print("â–ˆ" * 65)

final_knowledge = campus_kbs.run_forward_chaining()
campus_kbs.display_results()

print("\n" + "â–ˆ" * 65)
print("ğŸ¯ STEP 4: FINAL ACCESS DECISION")
print("â–ˆ" * 65)

if campus_kbs.check_fact_exists("campus_entry_approved"):
    print("\nâœ…âœ…âœ… FINAL DECISION: ENTRY GRANTED âœ…âœ…âœ…")
    if campus_kbs.check_fact_exists("research_lab_entry_permitted"):
        print("   â–¶ï¸ Access Level: RESEARCH PERSONNEL (Full Lab Access)")
    elif campus_kbs.check_fact_exists("temporary_access_granted"):
        print("   â–¶ï¸ Access Level: VISITOR (Temporary Limited Access)")
else:
    print("\nâŒâŒâŒ FINAL DECISION: ENTRY DENIED âŒâŒâŒ")
```
